```python
import pandas as pd
import numpy as np
```

```python
p4data = pd.read_csv('gdrive/My Drive/p4dataset2023.txt', header=None, delim_whitespace=True)
```

```python
newp4data = p4data.drop(p4data.columns[:3], axis=1)
```

```python
newp4data
```

```python
newp4data.reset_index(drop=True, inplace=True)
newp4data.columns = range(newp4data.shape[1])  # Reset column index to start from 0
```

```python
newp4data
```

```python
modes = newp4data.mode().iloc[0]
```

```python
for i in range(len(newp4data)):
    for j in range(len(newp4data.columns)):
        if newp4data.iat[i, j] == modes.iloc[j]:
          newp4data.iat[i, j] = 0
        else:
          newp4data.iat[i, j] = 1
```

```python
newp4data
```

```python
X = newp4data.copy()
X
```

# Exercise a)

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

```python
pca = PCA()
pca.fit(X)

ev = pca.components_
print("The dimension of the returned eigenbvectors is", ev.shape)
```

# Exercise b)

```python
pca = PCA(n_components = 2)
pca.fit(X)
components = pca.transform(X)
b1 = components.T[0]
b2 = components.T[1]
```

```python
population = p4data[2]
distinct_pop = set(population)
distinct_pop
```

```python
colours = []

for i in population:
  if i == 'ACB':
    colours.append('red')
  elif i == 'ASW':
    colours.append('blue')
  elif i == 'ESN':
    colours.append('yellow')
  elif i == 'GWD':
    colours.append('orange')
  elif i == 'LWK':
    colours.append('green')
  elif i == 'MSL':
    colours.append('purple')
  elif i == 'YRI':
    colours.append('cyan')
```

```python
#scatter plot

for a,b,x,y in zip(b1,b2,colours,population):
  plt.scatter(a, b, c = x, label = y)

plt.title('Projection of X onto the First Two Principal Components')
plt.xlabel('b1')
plt.ylabel('b2')

distinct_colours = [plt.Line2D([0], [0], marker='o', color='w', markerfacecolor='red'),
        plt.Line2D([0], [0], marker='o', color='w', markerfacecolor='blue'),
        plt.Line2D([0], [0], marker='o', color='w', markerfacecolor='yellow'),
        plt.Line2D([0], [0], marker='o', color='w', markerfacecolor='orange'),
        plt.Line2D([0], [0], marker='o', color='w', markerfacecolor='green'),
        plt.Line2D([0], [0], marker='o', color='w', markerfacecolor='purple'),
        plt.Line2D([0], [0], marker='o', color='w', markerfacecolor='cyan')]

plt.legend(distinct_colours,distinct_pop)
plt.show()
```

# Exercise c)

The two facts that I oberserved viewing the above plot were that based on the 7 distinct population, they appear to come as clusters for each data point. Because of this observation, I can conclude there is some sort of pattern/correlation between the first 2 principal components. The first two principal components capturing the clusters of population tells me that due to the commnon plot on the projection of X, the genes have a great deal of resemblance when they are from the same population (history and geography) of location.

```python
absb3 = abs(pca.components_[2])
nuc_index = np.arange(1, 10102)

plt.scatter(nuc_index, absb3)

plt.title('Nucleobase Index vs. Absolute Value of Corresponding Third Principal Component')
plt.xlabel('Nucleobase Index')
plt.ylabel('Absolute Value of Corresponding Third Principal Component')
plt.show()
```

I notice that as we reach around the hgher (>9000) nucleobase index, I see the absolute value of  the third principal component becoming higher. This could because we concluded that the third principal component captures the difference in gender (male vs female), so after around the 9500th index of nucleobases, chromosomes has a bigger part to play, hence the bigger difference.
